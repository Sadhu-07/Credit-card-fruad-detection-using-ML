```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define pairs of features and their correlations
feature_pairs = [('V17', 'V18'), ('V16', 'V17'), ('V14', 'V4'), ('V12', 'V10'), ('V12', 'V11'), ('V11', 'V10'), ('V3', 'V10'), ('V3', 'V12'), ('V9', 'V10')]

# Set up the plot
fig, axes = plt.subplots(nrows=len(feature_pairs), ncols=1, figsize=(8, 6*len(feature_pairs)))

# Loop through each feature pair
for i, (feature1, feature2) in enumerate(feature_pairs):
    # Create scatter plot
    sns.scatterplot(x=feature1, y=feature2, hue='Class', data=df, ax=axes[i], alpha=0.5)
    
    # Add annotations for correlation
    correlation = df[feature1].corr(df[feature2])
    axes[i].annotate(f'Correlation: {correlation:.2f}', xy=(0.05, 0.95), xycoords='axes fraction', fontsize=12, ha='left', va='top')
    
    # Add titles and labels
    axes[i].set_title(f'Scatter Plot for {feature1} vs {feature2}')
    axes[i].set_xlabel(feature1)
    axes[i].set_ylabel(feature2)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
#spliting the data
```

```python
x = df.drop(['id','Class'],axis=1)
y = df.Class
```

```python
x.head()
```

```python
print('Shape of x',x.shape)
print('Shape of y',y.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=15,stratify= y)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#logistic regression
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler  # Import StandardScaler for scaling the data

# Instantiate StandardScaler
scaler = StandardScaler()

# Fit and transform the training data
x_train_scaled = scaler.fit_transform(x_train)

# Instantiate Logistic Regression with a higher max_iter value
lr = LogisticRegression(max_iter=1000)

# Fit the logistic regression model on the scaled training data
lr.fit(x_train_scaled, y_train)

# Make predictions on the scaled training data
preds_lr_train = lr.predict(x_train_scaled)

# Evaluate the model on the training dataset
print('-------Training Accuracy---------')
def model_eval(actual, predicted):
    acc_score = accuracy_score(actual, predicted)
    conf_matrix = confusion_matrix(actual, predicted)
    clas_rep = classification_report(actual, predicted)
    print('Model Accuracy is: ', round(acc_score, 2))
    print(conf_matrix)
    print(clas_rep)
    
model_eval(y_train, preds_lr_train)
```

```python
# Plot feature importances
def plot_feature_importance(features, importances):
    feature_importance = pd.DataFrame({'Feature': features, 'Importance': importances})
    feature_importance = feature_importance.sort_values(by='Importance', ascending=False)
    
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Importance', y='Feature', data=feature_importance, palette='viridis')
    plt.xlabel('Importance')
    plt.ylabel('Feature')
    plt.title('Feature Importance')
    plt.show()

# Get feature importances from the logistic regression model
feature_importances = lr.coef_[0]
features = x_train.columns

# Plot feature importances
plot_feature_importance(features, feature_importances)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Instantiate Decision Tree Classifier
dt_classifier = DecisionTreeClassifier()

# Train the model
dt_classifier.fit(x_train, y_train)

# Make predictions on the test data
preds_dt_test = dt_classifier.predict(x_test)

# Evaluate the model on the test dataset
print('-------Test Accuracy---------')
model_eval(y_test, preds_dt_test)

# Plot the decision tree
plt.figure(figsize=(20, 10))
plot_tree(dt_classifier, feature_names=x_train.columns, class_names=['0', '1'], filled=True)
plt.show()
```

```python
# Get feature importances from the decision tree model
feature_importances_dt = dt_classifier.feature_importances_

# Plot feature importances
plot_feature_importance(features, feature_importances_dt)
```

```python
import xgboost as xgb

# Instantiate XGBoost Classifier
xgb_classifier = xgb.XGBClassifier()

# Train the model
xgb_classifier.fit(x_train, y_train)

# Make predictions on the test data
preds_xgb_test = xgb_classifier.predict(x_test)

# Evaluate the model on the test dataset
print('-------Test Accuracy---------')
print('Model Accuracy is: ', round(accuracy_score(y_test, preds_xgb_test), 2))
print(confusion_matrix(y_test, preds_xgb_test))
print(classification_report(y_test, preds_xgb_test))
```

```python
importance_type = "weight"  # or "gain" or "cover" or "total_gain" or "total_cover"
feature_importances_xgb = xgb_classifier.get_booster().get_score(importance_type=importance_type)

# Sort feature importances by their values
feature_importances_xgb = sorted(feature_importances_xgb.items(), key=lambda x: x[1], reverse=True)

# Plot feature importances
plt.figure(figsize=(10, 6))
plt.barh([x[0] for x in feature_importances_xgb], [x[1] for x in feature_importances_xgb])
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importance (XGBoost)')
plt.show()
```

```python
import lightgbm as lgb

# Convert the dataset into LightGBM format
lgb_train = lgb.Dataset(x_train, label=y_train)
lgb_test = lgb.Dataset(x_test, label=y_test)

# Set parameters for LightGBM
params = {
    'objective': 'binary',
    'metric': 'binary_error'
}

# Train the model
num_round = 100
lgb_model = lgb.train(params, lgb_train, num_round, valid_sets=[lgb_train, lgb_test])

# Make predictions on the test data
preds_lgb_test = lgb_model.predict(x_test, num_iteration=num_round)

# Convert predicted probabilities to binary predictions
preds_lgb_test_binary = [1 if pred > 0.5 else 0 for pred in preds_lgb_test]

# Evaluate the model on the test dataset
print('-------Test Accuracy---------')
print('Model Accuracy is: ', round(accuracy_score(y_test, preds_lgb_test_binary), 2))
print(confusion_matrix(y_test, preds_lgb_test_binary))
print(classification_report(y_test, preds_lgb_test_binary))

# Get feature importances from the LightGBM model
feature_importances_lgb = lgb_model.feature_importance(importance_type='gain')

# Plot feature importances
plot_feature_importance(features, feature_importances_lgb)
```
